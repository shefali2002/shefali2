```python
def remove_url(text):
    url_pattern = r'((http|ftp|https):\/\/)?[\w\-_]+(\.[\w\-_]+)+([\w\-\.,@?^=%&amp;:/~\+#]*[\w\-\@?^=%&amp;/~\+#])?'
    text = re.sub(url_pattern, ' ', text)
    #text=re.sub("?",'no',text)
    #text=re.sub('[0-9]hrs*','no',text)
    #text=re.sub('[0-9]hours*','no',text)
    #text=re.sub('[0-9] hrs*','no',text)
   # text=re.sub('[0-9] hours*','no',text)
   # text=re.sub('[0-9]hr',"",text)
   # text=re.sub('no.','okay',text)
    

    text=re.sub("[#!@$&/?\,.]","",text)
    text=re.sub("[0-9]","",text)
    text=re.sub("@[A-Z_a-z0-9]+","",text)   
    text=re.sub("#[A-Z_a-z0-9]+","",text)   
    #text=re.sub('[^A-Za-z]+', ' ', text)
    # standardize white space
    text = re.sub(r'\s+', ' ', text)
    #remove white space
    text = text.strip()
    return text
text1="he is 23563 # jdn उदाहरण"
print(remove_url(text1))
tpddl['clean_sentence']=tpddl['hit_sentence'].apply(lambda x:remove_url(x)) 
train['clean_sentence']=train['hit_sentence'].apply(lambda x:remove_url(x)) 
#print(tpddl['clean_sentence'])
tpddl['clean_sentence'].tail()
train['clean_sentence'].tail()
```

```python
df=pd.read_csv('hinglish.txt',names=['words'])
new_stopwords=df['words'].tolist()
negation_words = ["not","none", "no", "never", "neither", "never", "no one", "nobody","nor", "nothing", "nowhere","by","when","but","however","contrary","otherwise","nonetheless","whereas","or","while"]
new_stopwords=[word for word in new_stopwords if word not in negation_words]
stop_words=nltk.corpus.stopwords.words('english')

stop_words.extend(new_stopwords)
#print(stop_words)
def stopwords_remove(text):
    #print(text)
    words=nltk.word_tokenize(text)
    #print(words)
    words=[word for word in words if word not in stop_words]
    text=' '.join(words)
    return text    
tpddl['clean_sentence']=tpddl['clean_sentence'].apply(lambda x: stopwords_remove(x))  
train['clean_sentence']=train['clean_sentence'].apply(lambda x: stopwords_remove(x))
#print(tpddl['clean_sentence'])
```

```python
lemmatizer=WordNetLemmatizer()
def lemmatization_word(text):
    words=nltk.word_tokenize(text)
    words=[lemmatizer.lemmatize(word) for word in words]
    text=' '.join(words)
    return text
tpddl['clean_sentence']=tpddl['clean_sentence'].apply(lambda x: lemmatization_word(x))
train['clean_sentence']=train['clean_sentence'].apply(lambda x: lemmatization_word(x))
print(tpddl['clean_sentence'])
print(train['clean_sentence'])
```
